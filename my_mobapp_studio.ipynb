```python
import pandas as pd 
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def load_dataset():
    data_fm = pd.read_csv('C:/Users/user/Desktop/My_mob_app/googleplaystore.csv')
    data_fm.drop(['Last Updated',"Current Ver", "Android Ver"],axis=1, inplace=True)
    return data_fm
apps = load_dataset()
apps.head(20)
```

```python
def print_summarize_dataset(dataset):
    print("This is details about dataset")
    print(dataset.info())
    
    print("This is statistic facts about dataset",dataset.describe())
    print("This is figures of dataset:",dataset.shape)
```

```python
print_summarize_dataset(apps)
```

```python
def change_size_kb(row):
    if type(row) ==str:
        if row[-1].lower() == "k":
            row = row.replace('k', '')
            return float(row[:-1]) / 1024
        elif row[-1].lower() == "m":
            row = row.replace('M', '')
            return float(row[:-1])
        elif row.isalpha():
            return None
    return row
```

```python
def clean_dataset(data_st):
    data_st = data_st.drop_duplicates(subset=["App"], keep='last')
    data_st["App"] = data_st["App"].str.replace('&', '-')
    data_st = data_st[data_st['Category'] != '1.9']
    data_st["Category"]=data_st["Category"].str.replace("_"," ")
    data_st["Category"]= data_st["Category"].apply(lambda x: x.capitalize())
    filtration= {'Everyone 10+': 'Everyone','Adults only 18+': 'Adults only','"': '', ',': '', ';': ' ','Mature 17+': 'Mature', 'and up': '','Unrated': 'NaN','Varies with device': 'NaN'}
    data_st.replace(filtration, regex=True, inplace=True)
    data_st["Rating"]= data_st["Rating"].astype(float)
    data_st["Reviews"]= data_st["Reviews"].astype(float)
    data_st["Size"]= data_st["Size"].apply(change_size_kb)
    data_st['Size'].replace('Varies with device', np.nan, inplace=True)
    data_st['Size'] = pd.to_numeric(data_st['Size'], errors='coerce')
    data_st['Size'].fillna(data_st['Size'].mean(), inplace=True)
    data_st["Installs"]= data_st["Installs"].str.replace("+", "").str.replace(",","").astype(int)
    data_st.rename(columns={'Type': 'Payment'}, inplace=True)
    data_st["Price"] = data_st["Price"].str.replace('$','').astype(float)
    data_st['Content Rating'] = data_st['Content Rating'].str.replace(r'[+]', '', regex=True)
    data_st['Content Rating'] = data_st['Content Rating'].str.replace('NaN', 'Unrated')
    data_st["Genres"] = data_st["Genres"].str.replace("&",'-')
    data_st['Popularity'] = data_st.Rating *10**7  * data_st.Reviews *10**2 * data_st.Installs
    data_st.dropna(inplace=True)
    return data_st
```

```python
cleaned_apps = clean_dataset(apps)
```

```python
def print_histograms(data_st):
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=[ "Popularity of Genres", "Installs of Genres"],
                        column_widths=[1,1], row_heights=[0.5])
    fig.add_trace(go.Histogram(x=data_st['Genres'], y=data_st['Popularity']), row=1, col=1)
    fig.add_trace(go.Histogram(x=data_st['Genres'], y=data_st['Installs']), row=1, col=2)
    fig.update_layout(width=1300, height=600)
    fig.show()
    fig = px.histogram(cleaned_apps, x='Category', y='Rating', title='The rating  of  category',color='Rating',barmode='group',width=1200, height=600)
    fig.show()
```

```python
print_histograms(cleaned_apps)
```

```python
def compute_correlations_matrix(dataset):
    corr_dataset= dataset.corr( numeric_only=True)
    df= px.imshow(corr_dataset,title='Correlation of DataFrame',text_auto=True, width=1000,height=800,aspect= 'equal')
    df.show()
```

```python
compute_correlations_matrix(cleaned_apps)
```
